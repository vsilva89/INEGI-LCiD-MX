### Procesamiento para el conteo de sitios tur√≠sticos en cada una de las Comunas/Corregimientos

```python
import geopandas as gpd
```

```python
df = gpd.read_file("Corregimientos_o_comunas_consitiosTuristicos.shp")
```

```python
df = df.dropna(subset=['identifica', 'NUMPOINTS2'])
```

```python
df_sorted = df.sort_values(by='NUMPOINTS2',ascending=False)
```

```python
df_sorted[["nombre","identifica","NUMPOINTS2"]].to_csv("dataset_sitios_x_comuna_corregimiento.csv")
```

```python
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(10, 3))
plt.bar(df_sorted['identifica'], df_sorted['NUMPOINTS2'], color='blue')  # Puedes cambiar bar() a plot() para una l√≠nea
plt.xlabel('Comuna/Corregimiento')
plt.ylabel('Total de Actractivos Turisticos')
plt.title('Actractivos Turisticos Medellin')
plt.xticks(rotation=45)  # Para rotar las etiquetas del eje x si son largas
plt.show()
```
