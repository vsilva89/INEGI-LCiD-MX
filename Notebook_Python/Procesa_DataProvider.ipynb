### Procesamiento para el conteo de sitios web con enfoque turismo sotenible

```python
import pandas as pd
```

```python
df = pd.read_json("un_datathon_dataprovider.json")
```

```python
servicios_turismo_sostenible = [
    "Hotels, Rooming Houses, Camps, and other Lodging Places",
    "Amusement and Recreation Services",
    "Museums, Art Galleries, and Botanical and Zoological Gardens",
    "Agricultural Production - Crops",
    "Agricultural Production - Livestock",
    "Local and Suburban Transit and Interurban Highway Passenger Transportation",
    "Food Stores",
    "Eating and Drinking Places",
    "Building Materials, Hardware, Garden Supply, and Mobile Home Dealers",
    "Water Transportation",
    "Personal Services"
]
```

```python
df_filtrado = df[df['sicmajorgroup'].isin(servicios_turismo_sostenible)]
```

```python
df_agrupado = df_filtrado.groupby('sicmajorgroup')['domain'].count().reset_index()
df_agrupado.columns = ['Tipo de Servicio', 'Total de Sitios Web']
```

```python
df_agrupado.to_csv("dataset_dataprovider.csv")
```

```python
import matplotlib.pyplot as plt

# Crear la grfica de barras para los tipos de servicios y su nmero de sitios web.
plt.figure(figsize=(10, 4))
plt.barh(df_agrupado['Tipo de Servicio'], df_agrupado['Total de Sitios Web'], color='teal')
plt.xlabel('Total de Sitios Web')
plt.ylabel('Tipo de Servicio')
plt.title('Sitios Web por Tipo de Servicio')
plt.tight_layout()
plt.show()
```
